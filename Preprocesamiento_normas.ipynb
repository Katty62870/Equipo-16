```python
import PyPDF2

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        num_pages = len(reader.pages)
        for page_num in range(num_pages):
            page = reader.pages[page_num]
            text += page.extract_text()
    return text
```

```python
pdf_path = "/content/NMX-CH-036-SCFI-1994 INDICADORES DE CARATULA.pdf"
text = extract_text_from_pdf(pdf_path)
print(text)
```

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
sw = stopwords.words('spanish')
print(len(sw))
print(sw)
```

```python
def clean_tok(tokens):

    tokens = tokens.lower() # convierte todo el texto en minusculas
    tokens = tokens.translate(str.maketrans('.',' ')) # solucion a palabras unidas por .
    tokens = tokens.translate(str.maketrans('','',string.punctuation)) # depura signos de puntuacion y caracteres especiales
    tokens = nltk.word_tokenize(tokens) #genera los tokens
    tokens = [token for token in tokens if token not in sw] #Elimina las stopword de la lista
    tokens = [token for token in tokens if len(token) > 1] # Elimina tokens menores a 1 caracter

    return(tokens)
```

```python
#cleandf = [clean_tok(x) for x in dfx]
cleandf = clean_tok(dfx)
```

```python
wordlem = WordNetLemmatizer()

def clean_doc(doc):

    doc = [wordlem.lemmatize(i, 'v') for i in doc] # lematizacion verbos
    doc = [wordlem.lemmatize(i, 'n') for i in doc] # lematizacion pronombres
    doc = [wordlem.lemmatize(i, 'r') for i in doc] # lematizacion adverbios
    #doc = [w for w in doc if w.isalpha()] # no se tienen en cuenta #
    doc = list(set(doc)) # elimiacion duplicados
    doc = [token for token in doc if len(token) > 1] #se vuelve a depuran token menores a 1

    return doc
```
